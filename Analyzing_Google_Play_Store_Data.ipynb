```python
import pandas as pd
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("apps.csv")
```

```python
df.fillna(0,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated()
```

```python
df.drop_duplicates(inplace = True)
```

```python
df['Last Updated'] = pd.to_datetime(df['Last Updated'])

```

```python
df.head()
```

```python
df= df.replace(regex='Varies with device', value='0')

```

```python
df=df.replace(regex='and up', value='')
```

```python
df.shape
```

```python
df.info()
```

```python
df.head(10)
```

```python
df.describe()
```

```python
df.ndim
```

```python
df['Category'].nunique()
```

```python
df['Installs'].replace(regex=r'\+', value='',inplace=True)
```

```python
df['Installs']= df['Installs'].str.replace(',','')
```

```python
 df["Installs"] = df.Installs.astype(float)
```

```python
df.head(10)
```

```python
total_installs=df.groupby('Category')['Installs'].sum().reset_index()
print(total_installs)
```

```python
fig = px.bar(total_installs, x='Category', y='Installs')
fig.show()
```

```python
communication_apps = df[df['Category'] == 'COMMUNICATION']
print(communication_apps)
```

```python
communication_apps_ranked_sum = communication_apps.groupby('Rating')['Installs'].sum().reset_index().sort_values(by='Rating', ascending=False)
print(communication_apps_ranked_sum)
top_5_communication_apps=communication_apps_ranked_sum.head(5)
```

```python
print(top_5_communication_apps)
```

```python
df_Reviews=df.sort_values(by='Reviews',ascending=False).head()
fig = px.bar(df_Reviews, x='Category', y='Reviews')
fig.show()
```

```python
free_apps = df[df['Type'] == 'Free']
paid_apps = df[df['Type'] == 'Paid']

fig = plt.figure(figsize=(10, 5))

sns.barplot(x=['Free', 'Paid'], y=[len(free_apps), len(paid_apps)], palette='Set2')

plt.title('Number of Free vs Paid Apps')
plt.xlabel('Type')
plt.ylabel('Number of Apps')
plt.show()


```

```python
content_rating_counts = df['Content Rating'].value_counts()


fig, ax = plt.subplots()
ax.pie(content_rating_counts, labels=content_rating_counts.index, autopct="%1.1f%%")
```

```python

teen_apps = df[df['Content Rating'] == 'Teen']
teen_category_counts = teen_apps['Category'].value_counts()

fig = px.bar(teen_category_counts, x=teen_category_counts.index, y=teen_category_counts.values, title='Count of Teen Apps by Category')
fig.show()
```

```python
top_rated_category = df.groupby('Category')['Rating'].mean().sort_values(ascending=False).head(5)

fig = px.bar(top_rated_category, x=top_rated_category.index, y='Rating', title='Top Rated Category',color_discrete_sequence=["brown"],width=800,height=400)
plt.figure(figsize=(20,5))
fig.show()
```

```python
most_reviewed_app = df.sort_values(by='Reviews', ascending=False).head(5)

fig = px.bar(most_reviewed_app, x='App', y='Reviews', title='App with most reviews',)
fig.show()
```
